```python
!pip install google-generativeai --quiet

import os
from datetime import datetime
import google.generativeai as genai
# Step 1: Set API Key
genai.configure(api_key=userdata.get('GOOGLE_API_KEY'))
# Shared Gemini model
model = genai.GenerativeModel("gemini-1.5-flash-latest")

# Agent 1: Symptom Agent
def symptom_agent(input_text):
    print("\nðŸ¤– [Symptom Agent]")
    response = model.generate_content(f"Classify the severity of this symptom and list possible conditions: {input_text}")
    print(response.text)
    return response.text

# Agent 2: Scheduler Agent
def scheduler_agent(symptom_severity_summary):
    print("\nðŸ“… [Scheduler Agent]")
    response = model.generate_content(f"Based on this summary, suggest next steps for the patient: {symptom_severity_summary}")
    print(response.text)
    return response.text

# Agent 3: Follow-Up Agent
def followup_agent():
    print("\nðŸ§  [Follow-Up Agent]")
    response = model.generate_content("What follow-up questions should a nurse ask to understand the patientâ€™s condition better?")
    print(response.text)
    return response.text

# Agent 4: Logger Agent
def logger_agent(name, pid, input_text, output_text):
    print("\nðŸ“š [Logger Agent] Recording information...")
    log_entry = f"\nTime: {datetime.now()}\nPatient: {name} (ID: {pid})\nInput: {input_text}\nAI Summary: {output_text}\n{'='*50}\n"
    with open("multiagent_conversation_log.txt", "a") as f:
        f.write(log_entry)
# Get user details
name = input("Enter your name: ")
pid = input("Enter your Patient ID: ")
user_input = input("Please describe your health symptoms: ")

# Agent 1: Symptom Understanding
summary = symptom_agent(user_input)

# Agent 2: Scheduling or Escalation
next_steps = scheduler_agent(summary)

# Agent 3: Ask more clarifying questions
followups = followup_agent()

# Agent 4: Log the conversation
logger_agent(name, pid, user_input, summary + "\n\n" + next_steps + "\n\n" + followups)
```
